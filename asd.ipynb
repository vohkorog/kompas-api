```python
import matplotlib.pyplot as plt
import numpy as np

def draw_hexagon(edge_length):
    """
          
    
    Args:
        edge_length (float):   
    """
    #     ( )
    angles = np.linspace(0, 2 * np.pi, 7)  # 7    
    
    #  
    x_coords = []
    y_coords = []
    
    for angle in angles:
        x = edge_length * np.cos(angle)
        y = edge_length * np.sin(angle)
        x_coords.append(x)
        y_coords.append(y)
    
    # 
    plt.figure(figsize=(8, 8))
    plt.plot(x_coords, y_coords, 'b-', linewidth=2)
    plt.fill(x_coords, y_coords, alpha=0.3, color='lightblue')
    
    #  
    plt.grid(True, alpha=0.3)
    plt.axis('equal')
    plt.title(f'  ( : {edge_length})')
    plt.xlabel('X')
    plt.ylabel('Y')
    
    #  
    plt.plot(0, 0, 'ro', markersize=8, label='')
    plt.legend()
    
    plt.show()

#  ,         
def get_hexagon_coordinates(edge_length):
    """
       
    
    Args:
        edge_length (float):   
    
    Returns:
        tuple: (x_coords, y_coords) -  
    """
    angles = np.linspace(0, 2 * np.pi, 7)
    
    x_coords = edge_length * np.cos(angles)
    y_coords = edge_length * np.sin(angles)
    
    return x_coords, y_coords

#  
if __name__ == "__main__":
    #      5
    draw_hexagon(8)
    
    #       
    x, y = get_hexagon_coordinates(10)
    print("  :")
    for i, (xi, yi) in enumerate(zip(x, y)):
        print(f" {i}: ({xi:.2f}, {yi:.2f})")
        
    print(x[0], y[0])
    print(x[1], y[1])
    print(x[2], y[2])
```
